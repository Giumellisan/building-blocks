## Logistische Regression

```
temperature = seq(1,200,1)
ausfall = pmin(1,round(c(rep(0.75,50),rep(0.5,50),rep(0.25,50),
                         rep(-0.5,50)) 
                       + (runif(200,0,1))))
head(temperature)
head(ausfall)
summary(temperature)
summary(ausfall)
hist(ausfall)
```

```
plot(temperature, ausfall)
```

```
boxplot(temperature~ausfall)
```

#### Beispiel: Was ist die Wahrscheinlichkeit für Ausfall bei gegebener Temperatur?

```
# 1. Versuch: Ausfallwahrscheinlichkeit = a + b* Temperatur
plot(temperature, ausfall,ylim=c(0,2))
model = lm(ausfall~temperature)
abline(model)
```

```
fit = glm(ausfall ~ temperature , family=binomial ) # mit Maximum-Likelihood Schätzung
summary(fit)
```

```
estimated_probabilities = fit$fitted.values
```

```
plot(temperature, estimated_probabilities, type="l", lwd=4, col=2)
lines(temperature, ausfall, col=1, type ="p")
```

```
# Zeigt die Gerade, die wir gefittet haben
plot(temperature,log(estimated_probabilities/(1-estimated_probabilities)),type="l",lwd=4,col=2)
```
